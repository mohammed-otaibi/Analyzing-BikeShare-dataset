```python
import pandas as pd
import folium
from folium.plugins import HeatMap
from geopy.distance import geodesic
```

```python
# Filter data to exclude rows with zero coordinates
filtered_data = bikeshare_data[
    (bikeshare_data['start station latitude'] != 0) &
    (bikeshare_data['start station longitude'] != 0) &
    (bikeshare_data['end station latitude'] != 0) &
    (bikeshare_data['end station longitude'] != 0)
]
```

### ***Visualizing Most Used Stations on a Map in Nwe York city:***


```python
from IPython.display import display
print("\nVisualizing Most Used Stations on a Map in Nwe York city:")
# Prepare data for visualization
station_counts = filtered_data.groupby(['start station latitude', 'start station longitude']).size().reset_index(name='count')

# Create a Folium map
m = folium.Map(location=[station_counts['start station latitude'].mean(), 
                         station_counts['start station longitude'].mean()], zoom_start=12)

# Add stations as circles
for _, row in station_counts.iterrows():
    folium.CircleMarker(
        location=(row['start station latitude'], row['start station longitude']),
        radius=row['count'] / 1000,  # Scale the size of the circle by usage count
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(m)

# Display the map inline in Jupyter Notebook
display(m)  # This will render the map correctly in Jupyter Notebook
```

### ***Analyzing Distances Between Start and End Stations:***

```python
# Function to calculate distance between two coordinates
def calculate_distance(row):
    start_coords = (row['start station latitude'], row['start station longitude'])
    end_coords = (row['end station latitude'], row['end station longitude'])
    return geodesic(start_coords, end_coords).kilometers

# Apply the function to calculate distances
filtered_data['distance_km'] = filtered_data.apply(calculate_distance, axis=1)

# Summary statistics for distance
distance_stats = filtered_data['distance_km'].describe()
print("Distance Analysis (in kilometers):")
print(distance_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot a histogram of trip distances
plt.figure(figsize=(10, 6))
sns.histplot(filtered_data['distance_km'], bins=50, kde=True, color='blue')
plt.title('Distribution of Trip Distances', fontsize=16)
plt.xlabel('Distance (km)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
